```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(x='Type', y='Total Interactions', data=df)
plt.xlabel('Post Type')
plt.ylabel('Total Interactions')
plt.title('Total Interactions by Post Type')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Lifetime Post Total Reach', y='Lifetime Post Total Impressions', data=df)
plt.xlabel('Lifetime Post Total Reach')
plt.ylabel('Lifetime Post Total Impressions')
plt.title('Scatter Plot: Lifetime Post Total Reach vs. Lifetime Post Total Impressions')
plt.show()
```

```python
# Histogram: Lifetime Post Total Reach
plt.figure(figsize=(10, 6))
sns.histplot(df['Lifetime Post Total Reach'], bins=20, kde=False, color='lightcoral')
plt.xlabel('Lifetime Post Total Reach')
plt.ylabel('Frequency')
plt.title('Histogram: Lifetime Post Total Reach')
plt.show()
```
