```python
#replacing null values
df['bedrooms'].fillna(int(df['bedrooms'].mean()), inplace=True)
df['bathrooms'].fillna(int(df['bathrooms'].mean()), inplace=True)
df['stories'].fillna(int(df['stories'].mean()), inplace=True)
df['parking'].fillna(int(df['parking'].mean()), inplace=True)
```

```python
df=df.fillna(df.mode())
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df)
```

```python
df['airconditioning'].replace({'yes': '1'}, inplace=True)
df['airconditioning'].replace({'no': '0'}, inplace=True)
```

```python
df
```

```python
df['mainroad'].replace({'yes': '1','no':'0'}, inplace=True)
df['basement'].replace({'yes': '1','no':'0'}, inplace=True)
df['prefarea'].replace({'yes': '1','no':'0'}, inplace=True)
df['guestroom'].replace({'yes': '1','no':'0'}, inplace=True)
df['basement'].replace({'yes': '1','no':'0'}, inplace=True)
df['hotwaterheating'].replace({'yes': '1','no':'0'}, inplace=True)
df['furnishingstatus'].replace({'furnished': '1','unfurnished':'0', 'semi-furnished':'-1'}, inplace=True)
df
```

```python
#scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bedrooms', y='bathrooms', data=df)
plt.xlabel('bedrooms')
plt.ylabel('bathrooms')
plt.title('Scatter Plot: bedrooms vs. bathrooms')
plt.show()
```

```python
# Histogram: Lifetime Post Total Reach
plt.figure(figsize=(10, 6))
sns.histplot(df['bedrooms'], bins=20, kde=False, color='lightcoral')
plt.xlabel('bedrooms')
plt.ylabel('Frequency')
plt.title('Histogram: frequency of bedrooms')
plt.show()
```

```python
#Barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='furnishingstatus', y='bedrooms', data=df)
plt.xlabel('furnishingstatus')
plt.ylabel('bedrooms')
plt.title('furnishingstatus of bedrooms')
plt.show()
```
