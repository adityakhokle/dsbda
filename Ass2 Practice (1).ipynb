# Z Score

```python
# To remove outliers
```

```python
def remove_outliers_zscore(data, threshold=3):
    numeric_data = df.select_dtypes(include=[np.number])                # Select only numeric columns
    z_score = (numeric_data - numeric_data.mean()) / numeric_data.std()
    df_cleaned = df[ (np.abs(z_score) < threshold).all(axis=1) ]
    return df_cleaned

data_cleaned = remove_outliers_zscore(df)
data_cleaned
```

# MODEL BUILDING

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.svm import SVC
```

```python
x = df.drop(['Diagnosis'], axis=1)
y = df['Diagnosis']
```

```python
y = y.astype("int")
```

```python
x_train, x_test, y_train, y_test = tts(x, y, train_size=0.75)
```

# MODEL TRAINING

```python
svc_model = SVC(random_state=0)
svc_model.fit(x_train, y_train)
```

```python
y_pred = svc_model.predict(x_test)
y_pred
```
